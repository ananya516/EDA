```python
#EDA 1

import pandas as pd

# Corrected path (either raw string or double backslashes)
df = pd.read_csv(r"/content/BIKE DETAILS.csv")

# Preview the dataset
df.head()
```

```python
price_range = df['selling_price'].max() - df['selling_price'].min()
print("Range of selling prices:", price_range)
```

```python
median_price = df['selling_price'].median()
print("Median selling price:", median_price)

```

```python
common_seller = df['seller_type'].mode()[0]
print("Most common seller type:", common_seller)
```

```python
count_50k = df[df['km_driven'] > 50000].shape[0]
print("Bikes with >50,000 km driven:", count_50k)
```

```python
avg_km_by_owner = df.groupby('owner')['km_driven'].mean()
print(avg_km_by_owner)
```

```python
prop_2015_or_older = (df[df['year'] <= 2015].shape[0]) / df.shape[0]
print("Proportion of bikes from 2015 or older:", round(prop_2015_or_older * 100, 2), "%")
```

```python
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
seller_counts = df['seller_type'].value_counts()
print(seller_counts)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

first_owner_df = df[df['owner'] == '1st owner']
sns.scatterplot(data=first_owner_df, x='km_driven', y='selling_price')
plt.title('Selling Price vs. KM Driven (1st Owner)')
plt.show()
```

```python
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

filtered_df = df[(df['km_driven'] >= Q1 - 1.5*IQR) & (df['km_driven'] <= Q3 + 1.5*IQR)]
print("Shape before:", df.shape)
print("Shape after removing outliers:", filtered_df.shape)
```

```python
sns.boxplot(data=df, x='year', y='selling_price')
plt.xticks(rotation=45)
plt.title('Selling Price by Year')
plt.show()
```

```python
from datetime import datetime

current_year = datetime.now().year
df['age'] = current_year - df['year']
df['depreciation'] = df['ex_showroom_price'] - df['selling_price']

avg_depreciation = df.groupby('age')['depreciation'].mean()
print("Average depreciation by bike age:\n", avg_depreciation)
```

```python
df['mean_price_by_year'] = df.groupby('year')['selling_price'].transform('mean')
above_avg_bikes = df[df['selling_price'] > 1.5 * df['mean_price_by_year']]
print(above_avg_bikes[['name', 'year', 'selling_price', 'mean_price_by_year']])
```

```python
numeric_cols = df.select_dtypes(include='number')
correlation_matrix = numeric_cols.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# EDA 4

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 📥 Step 1: Load dataset (adjust filename/path as needed)
df = pd.read_csv("/content/spotify.csv")  # replace with your actual filename

# Step 2: Data quality checks
print("Shape:", df.shape)
print("Null values:\n", df.isnull().sum())
print("Duplicate rows count:", df.duplicated().sum())

# 💧 Handling nulls & duplicates
df = df.drop_duplicates()
df = df.dropna(subset=['Artist', 'Track Name', 'Popularity', 'Duration (ms)'])

# Step 3: Distribution of popularity (Histogram)
plt.figure(figsize=(8,5))
sns.histplot(df['Popularity'], bins=30, kde=True)
plt.title("Popularity Distribution")
plt.xlabel("Popularity")
plt.ylabel("Count")
plt.show()

# Step 4: Relationship between popularity and duration (Scatter plot)
plt.figure(figsize=(8,5))
sns.scatterplot(x='Duration (ms)', y='Popularity', data=df, alpha=0.6)
plt.title("Popularity vs Duration")
plt.xlabel("Duration (ms)")
plt.ylabel("Popularity")
plt.show()

# Step 5: Artist with most tracks (Countplot)
plt.figure(figsize=(10,5))
sns.countplot(y='Artist', data=df, order=df['Artist'].value_counts().index[:10])
plt.title("Top 10 Artists by Track Count")
plt.xlabel("Number of Tracks")
plt.ylabel("Artist")
plt.show()

# Step 6: Top 5 least popular tracks
least5 = df.nsmallest(5, 'Popularity')[['Artist', 'Track Name', 'Popularity']]
print("Top 5 least popular tracks:\n", least5)

# Step 7: Top 5 most popular artists (by avg popularity)
artist_pop = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False)
top5_artists = artist_pop.head(5)
print("\nTop 5 artists by average popularity:\n", top5_artists)

# Step 8: For these top 5 artists, their most popular track
top_artists_tracks = {}
for artist in top5_artists.index:
    subset = df[df['Artist'] == artist]
    best = subset.loc[subset['Popularity'].idxmax()]
    top_artists_tracks[artist] = best['Track Name']
print("\nMost popular track per top artist:\n", top_artists_tracks)

# Step 9: Pairplot among numeric columns (Popularity and Duration)
sns.pairplot(df[['Popularity', 'Duration (ms)']])
plt.show()

# Step 10: Duration variation across artists (boxplot)
plt.figure(figsize=(12,6))
sns.boxplot(x='Artist', y='Duration (ms)', data=df[df['Artist'].isin(top5_artists.index)])
plt.title("Duration Distribution for Top Artists")
plt.xticks(rotation=45)
plt.show()

# Step 11: Popularity distribution across artists (violin plot)
plt.figure(figsize=(12,6))
sns.violinplot(x='Artist', y='Popularity', data=df[df['Artist'].isin(top5_artists.index)])
plt.title("Popularity Distribution for Top Artists")
plt.xticks(rotation=45)
plt.show()
```
